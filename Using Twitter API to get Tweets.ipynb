```python
usernames = list(user_data.screen_name)

def split(a, n):
    k, m = divmod(len(a), n)
    return (a[i*k+min(i, m):(i+1)*k+min(i+1, m)] for i in range(n))

usernames = list(split(usernames, 14))
usernames
```

```python
def clean_text(text):
    text = str(text).encode('utf-8').decode('ascii', 'ignore')
    text = re.sub("@[A-Za-z0-9]+","",text) #replace mention with "" 
    text = re.sub(r"http\S+", "", text) #replace http
    text = re.sub("\n+|\r+", " ", text) #replace escape character one space
    text = re.sub("\s\s+", " ", text) #repleace whitespace with one space
    text = re.sub('^\s', '', text) #replace whitespace in the beginning
    text = re.sub('\s$', '', text) #replace whitespace in the end
    return text

def get_user_timeline(user_id):
    userdict = {}
    placedict = {}
    tweets = []
    for result in tweepy.Paginator(client.get_users_tweets, 
                                   id  = user_id,
                                   end_time = '2022-01-01T00:00:01Z',
                                   exclude = ['retweets'],
                                   max_results = 100, 
                                   tweet_fields=['id', 'text', 'created_at', 'author_id', 'lang', 'geo'],
                                   user_fields=['username', 'name', 'location', 'description'], 
                                   place_fields=['id','name', 'full_name', 'country_code', 'country', 'place_type', 'geo'], 
                                   expansions='author_id,geo.place_id,referenced_tweets.id'):    

            if ('users' in result.includes.keys()):
                for user in result.includes['users']:
                    userdict[user.id] = {'username': user.username, 'name':user.name, 
                                         'location': user.location, 'description':user.description,
                                         'entities':user.entities}

            if ('places' in result.includes.keys()):
                for place in result.includes['places']:
                    placedict[place.id] = {'name': place.name, 'full_name': place.full_name, 'country_code': place.country_code, 'country': place.country, 'place_type': place.place_type, 'geo': place.geo}
            
            if result.data is None:
                continue
            
            for tweet in result.data:
                tweet_id = tweet.id
                text = clean_text(tweet.text)

                created_at = tweet.created_at
                author_id = tweet.author_id
                lang = tweet.lang
                name = ""
                location = ""
                description = ""
                pl_id = ""
                pl_name = ""
                pl_full_name = ""
                pl_country_code = ""
                pl_country = ""
                pl_place_type = ""
                pl_geo_type = ""
                pl_geo_bbox = ""

                author_info=userdict[tweet.author_id] 

                if (author_info is not None): 
                    location = author_info['location']
                    location = clean_text(location)
                    
                    description = author_info['description']
                    description = clean_text(description)

                placeid = tweet.geo 
                if (placeid is not None): 
                    if('place_id' in placeid.keys()):
                        pl_id = placeid["place_id"]
                        if (pl_id in placedict.keys()):   
                            pl=placedict[pl_id]
                            pl_name = pl["name"].encode('utf-8').decode('ascii', 'ignore')
                            pl_country_code = place.country_code
                            pl_country = place.country
                            pl_place_type = place.place_type
                            pl_country_code=pl["country_code"].encode('utf-8').decode('ascii', 'ignore')
                            pl_country=pl["country"].encode('utf-8').decode('ascii', 'ignore')
                            pl_place_type=pl["place_type"].encode('utf-8').decode('ascii', 'ignore')
                            pl_geo_type=pl["geo"]["type"].encode('utf-8').decode('ascii', 'ignore')
                            pl_geo_bbox=str(pl["geo"]["bbox"])   

                data = [tweet_id, created_at, 
                        author_id, author_info['username'], author_info['name'], location, description, 
                        text, lang, 
                        pl_id, pl_name, pl_full_name, pl_country_code, pl_country, pl_place_type, pl_geo_type, pl_geo_bbox]
                tweets.append(data)
    cols = ['tweet_id', 'created_at',' author_id', 'username', 'display_name', 'location', 'description',
                   'text', 'lang', 'pl_id', 'pl_name', 'pl_full_name', 
                   'pl_country_code', 'pl_country', 'pl_place_type', 'pl_geo_type', 'pl_geo_bbox']     
    df = pd.DataFrame(tweets, columns=cols)
    return df 
```

```python
tweets_df = pd.DataFrame()
count = 0 
for user_id, username in user_ids.items():
    count += 1 
    percentage = 100*count/len(user_ids)
    percentage =  "{:.2f}".format(percentage)
    print(count,'of',len(user_ids),' ',percentage,'%')
    print(user_id, username)
    data = get_user_timeline(user_id)
    print('test')
    tweets_df = pd.concat([tweets_df,data])
```
